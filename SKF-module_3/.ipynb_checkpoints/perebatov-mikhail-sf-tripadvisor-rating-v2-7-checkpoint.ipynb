# 3. Обработка признаков

```python
# Упростим представление списка кухонь, удалив лишние символы
data['Cuisine Style'] = data['Cuisine Style'].apply(lambda x: clean_string(x))
#(italian, mediterranean, european, vegetarian)

# Упростим представление отзывов, удалив лишние символы
data['Reviews'] = data['Reviews'].apply(lambda x: clean_string(x))
# (better than the links, ivy black, 12/28/2017,)

# приведем уровень цен к числам
price_dict = {
    '$': 1,
    '$$ - $$$': 2,
    '$$$$': 3}
# нормализуем поле Ranking, учитывая разное число ресторанов в городах.
data['Price Range'] = data['Price Range'].map(price_dict)

# уберем самые сильные выбросы в поле 'Number of Reviews'
data['Number of Reviews'] = data['Number of Reviews'].apply(
    lambda x: 2200 if x > 2200 else x)

# нормализуем поле Ranking, учитывая разное число ресторанов в городах.
# создадим словарь {город:макс. рейтинг}
City_ranking = data.groupby('City')['Ranking'].max().to_dict()
data['Ranking_NORM'] = data[['City', 'Ranking']].apply(
    lambda x: x[1]/City_ranking[x[0]], axis=1)
```

# 4. Добавление новых признаков

```python
# добавим поле дат последних отзывов
data['date_review'] = data['Reviews'].apply(lambda x: date_review(x))

# # добавим дату последнего отзыва
data['date_last_review'] = data['date_review'].apply(lambda x: x.max())

# # добавим поле года последнего отзыва
data['year_last_review'] = data["date_last_review"].dt.year

# добавим поле количество дней между отзывами
data['date_int_review'] = data['date_review'].apply(lambda x: x.max()-x.min())
data['date_int_review'] = pd.to_numeric(
    data['date_int_review'].dt.days, downcast='integer')

# добавим поле количества вида кухонь
data['total_cuisin'] = data['Cuisine Style'].apply(lambda x: len(x.split(',')))

# добавим поле популярности (встречаемости в мире) вида кухони в ресторанах
# создадим словарь встречаемости {кухня: встречаемость в dateset}
most_popular_cuisine = dict(
    Counter(data['Cuisine Style'].str.cat(sep=',').split(",")).most_common())
del most_popular_cuisine['other']
# присвоим c помощью словаря первую попавшуюся кухню значением распространённости
data['popular_cuisin'] = data['Cuisine Style'].apply(
    lambda x: replace_popular_cuisine(x) if x != 'other' else None)

# Создадим поле население города в тыс.чел
# использум словарь population представленный выше
data['Population'] = data['City'].map(population)

# Посчитаем отношение количества отзывов к населению:
data['Review to Population'] = data['Number of Reviews'] \
    / data['Population']

# Отношение нормализованного ранга к населению:
data['Rel_Rank to Population'] = data['Ranking_NORM'] \
    / data['Population']


# добавим поле хороший(1) или нейтральный, плохой отзыв(0)
dict_good = 'good|best|wonderful|unique|nice|amazing|excellent|great|\
lovely|healthy|delicious|traditional|historic|friendly|fresh|cheap|liked|heavenly|fast|not bad'

data['good_review'] = data['Reviews'].str.contains(dict_good)
data['good_review'] *= 1
```

# Data Preprocessing
Теперь, для удобства и воспроизводимости кода, завернем всю обработку в одну большую функцию.

```python
df_train = pd.read_csv(DATA_DIR+'/main_task.csv')
df_test = pd.read_csv(DATA_DIR+'kaggle_task.csv')
df_train['sample'] = 1  # помечаем где у нас трейн
df_test['sample'] = 0  # помечаем где у нас тест
# в тесте у нас нет значения Rating, мы его должны предсказать, по этому пока просто заполняем нулями
df_test['Rating'] = 0
data = df_test.append(df_train, sort=False).reset_index(
    drop=True)  # объединяем


def preparation(data):

    # заменим '[[], []]' на NaN  в 'Reviews'
    data['Reviews'] = data['Reviews'].apply(
        lambda x: None if x == '[[], []]' else x)

    # Для полей  Price Range, Cuisine Style, Number of Reviews вынесем информацию о наличии пропуска как отдельный признак.
    for column in data.loc[:, data.isna().any()].columns:  # список полей содержащий NAN
        data[column+'_NA'] = column_NAN(column)  # def

    # заменим NaN для пропусков в 'Number of Reviews' на 0 как наиболее логичное значение
    data['Number of Reviews'].fillna(0, inplace=True)

    # Упростим представление списка кухонь, удалив лишние символы
    data['Cuisine Style'] = data['Cuisine Style'].apply(
        lambda x: clean_string(x))
    # (European,Scandinavian,Danish,VegetarianFriendly,GlutenFreeOptions)

    # Упростим представление списка кухонь, удалив лишние символы
    data['Reviews'] = data['Reviews'].apply(lambda x: clean_string(x))

    # приведем уровень цен к числам
    price_dict = {
        '$': 1,
        '$$ - $$$': 2,
        '$$$$': 3}

    data['Price Range'] = data['Price Range'].map(price_dict)

#     # нормализуем поле Ranking, учитывая разное число ресторанов в городах.
    City_ranking = data.groupby('City')['Ranking'].max().to_dict()
    data['Ranking_NORM'] = data[['City', 'Ranking']].apply(
        lambda x: x[1]/City_ranking[x[0]], axis=1)

    # уберем самые сильные выбросы в поле 'Number of Reviews'
    data['Number of Reviews'] = data['Number of Reviews'].apply(
        lambda x: 2200 if x > 2200 else x)

    # добавим поле дат последних отзывов
    data['date_review'] = data['Reviews'].apply(lambda x: date_review(x))

    # # добавим дату последнего отзыва
    data['date_last_review'] = data['date_review'].apply(lambda x: x.max())

    # # добавим поле года последнего отзыва
    data['year_last_review'] = data["date_last_review"].dt.year

    # добавим поле количество дней между отзывами
    data['date_int_review'] = data['date_review'].apply(
        lambda x: x.max()-x.min())
    data['date_int_review'] = pd.to_numeric(
        data['date_int_review'].dt.days, downcast='integer')

    # добавим поле количества кухонь
    data['total_cuisin'] = data['Cuisine Style'].apply(
        lambda x: len(x.split(',')))

    # добавим поле популярности (встречаемости в мире) вида кухни в ресторанах
    # создадим словарь встречаемости кухни в dateset
    most_popular_cuisine = dict(
        Counter(data['Cuisine Style'].str.cat(sep=',').split(",")).most_common())
    del most_popular_cuisine['other']

    data['popular_cuisin'] = data['Cuisine Style'].apply(
        lambda x: replace_popular_cuisine(x) if x != 'other' else None)

    # добавим поле хороший (1) или нейтральный, плохой отзыв(0)
    dict_good = 'good|best|wonderful|unique|nice|amazing|excellent|great|\
    lovely|healthy|delicious|traditional|historic|friendly|fresh|cheap|liked|heavenly|fast|not bad'

    data['good_review'] = data['Reviews'].str.contains(dict_good)
    data['good_review'] *= 1

    # Создадим поле население города в тыс.чел
    # использум словарь Population представленный выше
    data['Population'] = data['City'].map(population)

    # Посчитаем отношение количества отзывов к населению:
    data['Review to Population'] = data['Number of Reviews'] \
        / data['Population']

    # Отношение нормализованного ранга к населению:
    data['Rel_Rank to Population'] = data['Ranking_NORM'] \
        / data['Population']
    
    # ЗАПОЛНИМ ПРОПУСКИ
    # заполним пропуски  медианой
    for column in ['total_cuisin', 'popular_cuisin', 'year_last_review', 'date_int_review', 'Price Range']:
        data[column].fillna(data[column].median(), inplace=True)

    # ранг ресторана на основании кол-ва отзывов
    data['ranking_rev_count'] = data['Ranking'] * data['Number of Reviews']

    # get_dummies
    data['City_name'] = data['City']
    data = pd.get_dummies(data, columns=['City'], dummy_na=False)

    # Добавим признак - индекс покупательской способности согласно рейтингу Numbeo (2018)

    data['purchasing_power'] = data['City_name'].map(purchasing_power_index)

    # убираем признаки типов "object" и "Datetime"
    columns_list = list(data.select_dtypes(include='number').columns)
    # убираем признаки ухудшающие модель
    columns_list.remove('Ranking')
    columns_list.remove('Number of Reviews_NA')
    columns_list.remove('Reviews_NA')
    # columns_list.remove('Ranking_NORM')
#     columns_list.remove('popular_cuisin')
    data = data[columns_list]
    return data


data = preparation(data)
```

```python
# Создаём модель (НАСТРОЙКИ НЕ ТРОГАЕМ)
model = RandomForestRegressor(n_estimators=100, verbose=1, n_jobs=-1, random_state=RANDOM_SEED)
```

```python
# Обучаем модель на тестовом наборе данных
model.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = model.predict(X_test)
```
